```python
%load_ext tensorboard
import tensorflow as tf

# Prepare Data for OR.
data = [
        [0, 0, 0], # Input 0 0 Output=> 0
        [1, 0, 1], # Input 1 0 Output=> 1
        [0, 1, 1], # Input 0 1 Output=> 1
        [1, 1, 1]  # Input 1 1 Output=> 1
]

# Import Pandas to make data in proper format
import pandas as pd

data =  pd.DataFrame(data)
x_input = data[[0,1]]
y_input = data[2]
```

```python
# Prepare tensorflow model

model = tf.keras.models.Sequential([
  tf.keras.Input(shape=(2,)),
  tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(
    optimizer= tf.compat.v1.train.GradientDescentOptimizer(0.5),
    loss=tf.keras.losses.MeanSquaredError()
)
```

```python
# Define the Keras TensorBoard callback.
from datetime import datetime

logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)

# Train the model
model.fit(x_input,y_input, epochs=500, callbacks=[tensorboard_callback])
```

```python
predictions = model.predict(x_input)
print(predictions)
%tensorboard --logdir logs
```
